### `Histogram`

  >Whats An Histogram ?

  - An Histogram Is A Type Of Graph That  Shows The Distribution of Numerical Data.

  - It Divides The Values Into Bins/Intervals .
  
  - Later We Can Count How Many Bins Fall Into Each Bin.


 



  ` Example`
- Lets Say We Collected Age Of `10 People`

  - `[22, 23, 25, 27, 28, 30, 32, 35, 36, 40]`

**Lets Say We Choose Bin Size Of 5 Years**
- 20-24
- 25-29
- 30-34
- 34-39
- 40-44


**Now lets Count Number Of People in Each Bin**
- 20-24 `(22,23)` -----> 2
- 25-29 `(25,27,28)` -----> 3
- 30-34 `(30,32)` -----> 2
- 34-39 `(35,36)` -----> 2
- 40-44`(40)` -----> 1




` Now We Plot histogram`

```python
import matplotlib.pyplot as plt 

Age = [22, 23, 25, 27, 28, 30, 32, 35, 36, 40]
```

```python
plt.hist(Age,bins=[20,25,30,35,40],color=['red'],edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.show()
```
